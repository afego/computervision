```python
# used_codes = list(set(df_even.code.unique() + df_2k.code.unique() + df_4k.code.unique())) # Doesnt work, the lists need to have same length

used_codes = set()
temp_list = []
for code in [df_even.code.unique(), df_2k.code.unique(), df_4k.code.unique()]:
    temp_list.extend(val for val in code if val not in used_codes)
    used_codes.update(code)
    
print(len(used_codes))
```

```python
df_images = pd.read_csv('data/images.csv')
print(df_images.shape)
df_images.head(3)
```

```python
df_images_codes = df_images.code.unique()

test_codes = [code for code in df_images_codes if code not in used_codes]
print(len(test_codes))
```

```python
df_test = df_images[df_images['code'].isin(test_codes)]
print(df_test.shape)
df_test.head()
```

```python
df_test['flood'] = df_test['tags'].apply(lambda x: 1 if len(x) > 2 else 0)
print(df_test.shape)
df_test.head()
```

```python
df_test = df_test.sample(frac=1, replace=False, random_state=42)
print(df_test.shape)
df_test.head()
```

The cameras not used in any training are extremaly disbalanced. I will pick at most 100 images for camera for flood = 1 (cameras 262, 414, 1162 and 3323), totalling 281 images. Then, I will choose the same number os images for class 0 between all cameras not used in training.

547 or 557 removed class 1 images

```python
sorted_test = df_test.sort_values(by='code', ascending=False)
sorted_test.shape
```

```python
class_1_codes = [262.0, 414.0, 1162.0, 3323.0]

indexes_to_drop = []
for cls_code in class_1_codes:
    indexes = sorted_test[(sorted_test.code == cls_code) & (sorted_test.flood == 1)].index.tolist()
    if len(indexes) > 100:
        indexes_to_drop += indexes[100:]
sorted_test = sorted_test.drop(indexes_to_drop)
sorted_test.shape
```

```python
count_by_code = sorted_test.groupby('code')['flood'].value_counts().unstack(fill_value=0)
print(count_by_code.shape)
count_by_code
```

```python
cls_1_total = count_by_code[1].sum()
cls_1_total
```

```python
img_per_camera = int(cls_1_total/(count_by_code.shape[0] - count_by_code[count_by_code[0] == 0].shape[0]))
img_per_camera
```

```python
indexes_to_drop = []
cameras_with_cls_0 = sorted_test[sorted_test.flood == 0]['code'].unique()

for cls_code in cameras_with_cls_0:
    indexes = sorted_test[(sorted_test.code == cls_code) & (sorted_test.flood == 0)].index.tolist()
    indexes_to_drop += indexes[img_per_camera:]
sorted_test = sorted_test.drop(indexes_to_drop)
sorted_test.shape
```

```python
count_by_code = sorted_test.groupby('code')['flood'].value_counts().unstack(fill_value=0)
count_by_code
```

```python
import shutil

def copy_images_from_csv(csv_file, output_folder):
    
    base_path = 'E:\GitHub\octa-flood-images\data\images'
    df = pd.read_csv(csv_file)
    
    if not os.path.exists(output_folder): 
        os.makedirs(output_folder)
    
    cls_0_path = os.path.join(output_folder,'0')
    cls_1_path = os.path.join(output_folder,'1')
    
    if not os.path.exists(cls_0_path):
        os.makedirs(cls_0_path)
        
    if not os.path.exists(cls_1_path):
        os.makedirs(cls_1_path)   
    
    for filepath in df.file_path.to_list():
        original_filepath = os.path.join(base_path,filepath)          
        output_filepath = os.path.join(output_folder, str(int(df[df.file_path == filepath]['flood'].values[0])))
        shutil.copy(original_filepath, output_filepath)
```

```python
cf = 'data/temp_test.csv'
od = 'E:/Datasets/cor-splits/temp-test/test'
copy_images_from_csv(cf, od)
```
